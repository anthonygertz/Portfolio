```python
train = pd.read_csv('data/train.csv')
test = pd.read_csv('data/test.csv')
```

```python
pd.set_option('display.max_columns', 500)
```

```python
full_columns = train.columns.tolist()
```

```python
full_columns.remove('permeability')
```

```python
X = pd.DataFrame(train, columns = full_columns)

y = train['permeability']
```

```python
X_train, X_test, y_train, y_test = train_test_split(
    X, y, random_state = 8669, test_size = 0.30
)
```

```python
ct = ColumnTransformer(transformers = [
    ('yeo-johnson', PowerTransformer(), full_columns),
], remainder = 'passthrough')

pipe = Pipeline(steps = [
    ('transform', ct),
    ('scale', MinMaxScaler())
    #('svm'), svm())
])
```

```python
X_train = pipe.fit_transform(X_train)
X_test = pipe.transform(X_test)
```

```python
X_train.shape
```

```python
model = Sequential()

#model.add(Dense(64,activation=ks.layers.LeakyReLU(alpha = 0.01)))
model.add(Dense(128, activation = 'relu'))
model.add(Dense(128, activation = 'relu'))
model.add(Dense(128, activation = 'relu'))
model.add(Dense(128, activation = 'relu'))
model.add(Dense(128, activation = 'relu'))
model.add(Dense(128, activation = 'relu'))
model.add(Dense(1, activation='linear'))

model.compile(loss = 'mean_squared_error', optimizer = 'adam', metrics = ['mae'])
```

```python
es = EarlyStopping(monitor = 'loss', patience = 250, restore_best_weights = True)
```

```python
mc = ModelCheckpoint(filepath = 'best_model_3090.h5', monitor = 'loss', save_best_only=True)
```

```python
X_es_train, X_es_test, y_es_train, y_es_test = train_test_split(X_train, y_train, test_size = 0.10, random_state = 8669)
```

```python
with tf.device('/gpu:0'):
    history = model.fit(x = X_es_train,y = y_es_train.values,
          validation_data = (X_es_test, y_es_test),
          batch_size = 128,
          epochs = 1000,
          callbacks = [es, mc])
```
